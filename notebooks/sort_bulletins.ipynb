```python
import pandas as pd
df = pd.read_csv("data/cleaned_data/applicants.csv")
job_descriptions = list(df["Job Description"])
for job in sorted(job_descriptions):
    print(job)
```

```python
import re

bulletin_ids = []

job_descriptions_stripped = []
for job in job_descriptions:
    try:
        bulletin_ids.append(re.split(r'(^[^\d]+)', job)[2])
        job = re.split(r'(^[^\d]+)', job)[1]
        job = job.replace("(", "")
        job = job.replace("-", "")
        job = job.replace("'", "")
        job = job.strip()
    
    except IndexError:
        pass
    
    job_descriptions_stripped.append(job.upper())

for job in sorted(job_descriptions_stripped):
    print(job)
```

```python
labeled_bulletins = []
all_bulletins = []
for bulletin in os.listdir("data/cleaned_data/job_bulletins"):
    all_bulletins.append(bulletin)
    try:
        stripped = (re.split(r'(^[^\d]+)', bulletin)[1])
        stripped = stripped.replace("(", "")
        stripped = stripped.replace("-", "")
        stripped = stripped.replace("_", "")
        stripped = stripped.strip()
        stripped = stripped.upper()
        print(stripped)
        if stripped in job_descriptions_stripped:
            labeled_bulletins.append(bulletin)
        elif stripped == "SENIOR ELECTRICAL TEST TECHNICIAN" or stripped == "WORKERS COMPENSATION CLAIMS ASSISTANT":
            labeled_bulletins.append(bulletin)
    except IndexError:
        labeled_bulletins.append(bulletin)
    
```

```python
for bulletin in os.listdir("data/cleaned_data/job_bulletins"):
    try:
        with open(f"data/cleaned_data/job_bulletins/{bulletin}", "r") as in_file:
            text = in_file.read()

        if bulletin in labeled_bulletins:
            with open(f"data/cleaned_data/job_bulletins/labeled/{bulletin}", "w", encoding="utf-8") as out_file:
                out_file.write(text)
        else:
            with open(f"data/cleaned_data/job_bulletins/unlabeled/{bulletin}", "w", encoding="utf-8") as out_file:
                out_file.write(text)
    
    except PermissionError:
        print(bulletin)

    
```

```python
ids = []
for bulletin in all_bulletin_ids:
    bl = bulletin.split()
    try:
        ids.append(bl[0])
    except IndexError:
        ids.append(bl)
```

```python
for id in ids:
    if any(c.isalpha() for c in id):
        ids.remove(id)
    elif len(id) > 4:
        ids.remove(id) 
```

```python
ids_ints = []
for id in ids:
    try:
        ids_ints.append(int(id))
    except TypeError:
        print(id)
```
