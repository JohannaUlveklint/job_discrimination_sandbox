```python
file = 'data/cleaned_data/job_bulletins/311 DIRECTOR  9206 041814.txt'
```

```python
def file_check(file):
    vectorizer = CountVectorizer(stop_words='english')
    with open(file) as f:
        # Remove newlines
        content = f.read().replace('\n', ' ')
        for i in range(10):
            # Remove numbers
            content = content.replace(str(i), '')
    print(content)
    # Count words in the file
    X = vectorizer.fit_transform([content])
    df_bow = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
    #print(df_bow)
    
    # Read feminine and masculine words
    with open('data/cleaned_data/men-ad-words.txt') as f:
        masculine_words = f.read().splitlines()
    with open('data/cleaned_data/women-ad-words.txt') as f:
        feminine_words = f.read().splitlines()
    
    words_in_doc = set(df_bow.columns.values)
    #print(words_in_doc)
    
    masculine_words_in_doc = list(words_in_doc.intersection(masculine_words))
    feminine_words_in_doc = list(words_in_doc.intersection(feminine_words))
    return df_bow[masculine_words_in_doc], df_bow[feminine_words_in_doc]
```

```python
male_df, female_df = file_check(file) 
```
