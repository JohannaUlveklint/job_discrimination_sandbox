```python
# Following the tutorial on https://www.youtube.com/watch?v=kxeyoyrf2cM&ab_channel=PatrickLoeber
from collections import Counter

# Count unique words
def counter_word(text_col):
    count = Counter()
    for text in text_col.values:
        for word in text.split():
            count[word] += 1
    return count

counter = counter_word(df.Text)
```

```python
num_unique_words = len(counter)
```

```python
# Split dataset into training and validation set
train_size = int(df.shape[0] * 0.8)

train_df = df[:train_size]
val_df = df[train_size:]

# Split text and labels
train_sentences = train_df.Text.to_numpy()
train_labels = train_df.Label_int.to_numpy()
val_sentences = val_df.Text.to_numpy()
val_labels = val_df.Label_int.to_numpy()
```

```python
# Tokenize
from tensorflow.keras.preprocessing.text import Tokenizer

# Vectorize a text corpus by turning each text into a sequence of integers
tokenizer = Tokenizer(num_words=num_unique_words)
tokenizer.fit_on_texts(train_sentences)  # fit only to training
```

```python
# Each word has unique index
word_index = tokenizer.word_index
```

```python
word_index
```

```python
train_sequences = tokenizer.texts_to_sequences(train_sentences)
val_sequences = tokenizer.texts_to_sequences(val_sentences)
```

```python
# print(train_sentences[10:11])
# print(train_sequences[10:11])
```

```python
# Pad the sequences to have the same length
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Max number of words in a sequence
max_length = 1000

train_padded = pad_sequences(train_sequences, maxlen=max_length, padding="post", truncating="post")
val_padded = pad_sequences(val_sequences, maxlen=max_length, padding="post", truncating="post")
train_padded.shape, val_padded.shape
```

```python
def decode(sequence):
    return " ".join([reverse_word_index.get(idx, "?") for idx in sequence])
```

```python
# Create LSTM model
from tensorflow.keras import layers

# Embedding: https://www.tensorflow.org/tutorials/text/word_embeddings
# Turns positive integers (indexes) into dense vectors of fixed size (other approach could be one-hot-enocoding).

# Word embedding give us a way to yse an efficient, dense representation in which similar words have
# a similar encoding. Importanty, you do not have to specify this encoding by hand. An embedding is a 
# dense vector of floating pont values (the length of the vector is a parameter you specify).

model = keras.models.Sequential()
model.add(layers.Embedding(num_unique_words, 32, input_length=max_length))

# The layer will take as input an integer matrix of size (batch, input_length),
# and the largest integer (i.e. word index) in the input should be no larger than num_wirds (vocabulary_size).
# Now model.output_shape is (None, input_length, 32), where "None" is the batch dimension.

model.add(layers.LSTM(64, dropout=0.1, return_sequences=True))
model.add(layers.LSTM(64, dropout=0.1))
model.add(layers.Dense(1, activation="sigmoid"))  # not linear

model.summary()
```

```python
loss = keras.losses.CategoricalCrossentropy(from_logits=False)  # Try another loss 
optim = keras.optimizers.SGD(lr=0.001)  # Try another optimizer, not Adam or SGD (seems not to be the case)
metrics = ["accuracy"]

model.compile(loss=loss, optimizer=optim, metrics=metrics)
```

```python
print(val_labels.astype(np.int))
```

```python
model.fit(train_padded, train_labels.astype(np.int), epochs=100, validation_data=(val_padded, val_labels.astype(np.int)), verbose=2, batch_size=20)
```

```python
predictions = model.predict(train_padded)
```
