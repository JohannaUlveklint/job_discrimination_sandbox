```python
import textwrap

def shorten_lines(file_name):
    with open(file_name, "r") as in_file:
        lines = in_file.readlines()

    with open(f"{file_name[:31]}/readable{file_name[31:]}", "w", encoding="utf-8") as out_file:
        for line in lines:
            if len(line) > 160:
                out_file.write(textwrap.fill(line, width=160))
                out_file.write("\n")

            else:
                out_file.write(line)
```
