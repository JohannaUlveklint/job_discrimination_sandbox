```python
import pandas as pd

applicants = pd.read_csv("applicants.csv")
```

```python
applicants
```

```python
mostly_women = applicants[applicants["Female"] >= (applicants["Apps Received"] - applicants["Unknown_Gender"]) * 0.6]
mostly_women
```

```python
mostly_men = applicants[applicants["Male"] >= (applicants["Apps Received"] - applicants["Unknown_Gender"]) * 0.6]
mostly_men
```

```python
neutral = applicants[(applicants["Female"] < (applicants["Apps Received"] - applicants["Unknown_Gender"]) * 0.6) & (applicants["Female"] > (applicants["Apps Received"] - applicants["Unknown_Gender"]) * 0.4)]
neutral
```

```python
int_women_men = pd.merge(mostly_women, mostly_men, how ="inner")
int_women_men
```

```python
int_women_neutral = pd.merge(mostly_women, neutral, how="inner")
int_women_neutral

# Accountant: 59.93413831% with unknown, 61,9047619 without
# Administrator: 58.09716599 with unknown, 61,45610278 without
```

```python
int_men_neutral = pd.merge(mostly_men, neutral, how="inner")
int_men_neutral
```
